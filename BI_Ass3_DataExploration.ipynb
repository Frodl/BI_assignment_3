### Sorting the dataframe in descending order by the amount of user ratings and reset the index:

```python
df_sorted = df.sort_values(by=['usersrated'], ascending=False)
df_sorted = df_sorted.reset_index()
```

### Plotting the data and graphically determine interesting sub-group:

#### Now we will visually explore the number of ratings per instance

```python
plt.figure(figsize =(7, 4))
plt.plot(df_sorted['usersrated'], linestyle = 'dotted')
plt.axvline(x=1000, color='b', label='x=1000')
plt.axhline(y=2793, color='r', label='y=2793')
plt.xlabel("Index sorted descending by how many user ratings the games received.")
plt.ylabel("number of user ratings")
plt.legend()
plt.show()
```

#### The data points were sorted descending by their number of ratings and show that there is only very few number of games that were rated very often compared to the rest.

#### Based on this finding we decided to create a subgroup of 5% the data that has the highest number of user ratings.

#### We subdivided the data by a threshold of 2793 user ratings, which creates a data split of 95/5 and accordingly 19000/1000 instances.

#### This may help later to verify potential bias of the model

### Creating separate dataframes for the two subgroups:

```python
df_10 = df_sorted[df_sorted['usersrated']>2793]
df_90 = df_sorted[df_sorted['usersrated']<=2793]
```

### Plotting boxplots for the two subgroups to compare their distribution regarding complexity and ratings:

```python
data_1 = df_10['avgweight']
data_2 = df_90['avgweight']
data = [data_1, data_2]

fig = plt.figure(figsize =(3, 4))
 
# Creating axes instance
ax = fig.add_axes([0, 0, 1, 1])
 
# Creating plot
bp = ax.boxplot(data)
 
# x-axis labels
ax.set_xticklabels(['10% most rated instances', 'rest 90%'])
ax.set_ylabel('game complexity')

# show plot
plt.show()
```

```python
data_1 = df_10['average']
data_2 = df_90['average']
data = [data_1, data_2]
 
fig = plt.figure(figsize =(3, 4))
 
# Creating axes instance
ax = fig.add_axes([0, 0, 1, 1])
 
# Creating plot
bp = ax.boxplot(data)
 
# x-axis labels
ax.set_xticklabels(['10% most rated instances', 'rest 90%'])
ax.set_ylabel('user rating')

# show plot
plt.show()
```

#### Comparing the two different groups with boxplots confirms the logical assumption that games that are more discussed by users are likely to be more popular.
#### The boxplots also confirm again the correlation between complexity and ratings, since the 10% subset does not only include higher ratings but also higher complexity.
#### Both ratings and complexity of the 10% most rated games are a bit higher than for the rest.
